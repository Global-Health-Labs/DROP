```python
import pandas as pd
import os
import glob
from PlateAnalyzer import *

## locate folder with raw data files. This will only identify files with an .xlsx file type. 
## Load the Quantification Amplification Result.xlsx from the Biorad software for each plate
## you want to analyze. We recommend selecting the "no background subtraction" option for 
## optimal results. 

path = os.getcwd()
folder = path + "\\raw_data"
csv_files = glob.glob(os.path.join(path, "*.xlsx"))

## Files that were identified for analysis 
print(str(len(csv_files)) + ' .xlsx files were identified for analysis. The files are listed here: ')
print(' ')
print(csv_files)
```

```python
## loops through each excel file in folder to analyze, generate csv files and plot
for f in csv_files:

    file_name = f.split("\\")[-1]
    pa = PlateAnalyzer(filename=file_name)
    params_df, qoi_df, data_dict = pa.analyze()

    # Generates PDF of all curves in plate 
    data_keys = list(data_dict.keys())    
    plot_vars = make_plots(data_dict, file_name)

    # export qoi and params_df files to csv for further processing
    qoi_df.to_csv(file_name[:-5] + '_qoi.csv')
    params_df.to_csv(file_name[:-5] + '_params.csv')
```
